```python
import pandas as pd
import glob
files = glob.glob('*.xlsx')
final_df = pd.DataFrame()
for file in files:
    df = pd.read_excel(file)
    final_df = pd.concat([df,final_df])
final_df = final_df.drop('Unnamed: 0',axis=1)
final_df = final_df.dropna(subset=['text'])
print(final_df)
pro = final_df[['post_id', 'text', 'image_lowquality', 'likes', 'comments', 'shares', 'post_url', 'user_id','username','commenter_id',
       'commenter_name', 'comment_text']]
pro = pro.rename(columns= {'likes':'Number of Likes','comments':'Number of Comments',"shares":"Number of Shares",'comment_text':'Comments'})
pro['DataSource'] = 'Facebook'
```

```python
pro.columns
```

```python
import re
pro = pro[['Comments']]
pro['Comments'] = pro['Comments'].apply(lambda x:re.sub(r"[@#$%^""&*‚Äô(_.)-+=!?/|\\',;:]", '',x) if x==x else x)
pro['Comments'] = pro['Comments'].apply(lambda x:re.sub(r"[‚Äú‚Äô‚Äù]", '',x) if x==x else x)
```

```python
pro = pro.dropna()
```

```python
import nltk
import swifter
from rake_nltk import Rake
nltk.download('stopwords')
nltk.download('punkt')
```

```python
r = Rake()

def rakeprocessing(text):
    r.extract_keywords_from_text(text)
    return r.get_ranked_phrases_with_scores()
```

```python
pro['Rake'] = pro.Comments.apply(lambda x:rakeprocessing(x))
```

```python
base_df = pro[['Comments']]
```

```python
rke = pro.explode('Rake')
print(rke)
rke=rke.dropna(subset=['Rake'])
rke['Rake_text'] = rke.Rake.apply(lambda x:x[1])
rke['Rake_Score'] = rke.Rake.apply(lambda x:str(int(x[0])))

```

```python
rke
```

```python
rke = rke[rke['Rake_Score'].isin(['1','2','3','4'])]
```

```python
rke = rke.drop(['Rake'],axis=1)
```

```python
from transformers import pipeline
classifier = pipeline('sentiment-analysis')
def sentimentprocessing(text):
    try:
        return classifier(text)
    except Exception as e:
        print(e)
        return text
```

```python
import swifter
sen['Sentiment_Analysis'] = sen.Comments.swifter.apply(lambda x:sentimentprocessing(x))
```

```python
sen['Sentiment_Score'] = sen['Sentiment_Analysis'].apply(lambda x:x[0]['score'])
sen['Sentiment'] = sen['Sentiment_Analysis'].apply(lambda x:x[0]['label'])
```

```python
sen
```

```python
sen.to_excel("sentiment_facbook.xlsx")
```
