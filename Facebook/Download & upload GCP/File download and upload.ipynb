```python
import wget
def fileiploaddownload(url):
    image_url = url
    image_filename = wget.download(image_url)
    print('Image Successfully Downloaded: ', image_filename)
    imageBlob = bucket.blob(image_filename)
    imageBlob.upload_from_filename(image_filename)
    outputurl = imageBlob.public_url
    return outputurl
```
