```python
import pandas as pd
import glob
```

```python
files = glob.glob('*.xlsx')
final_df = pd.DataFrame()
for file in files:
    df = pd.read_excel(file)
    final_df = pd.concat([df,final_df])
final_df = final_df.drop('Unnamed: 0',axis=1)
final_df = final_df.dropna(subset=['text'])
print(final_df)
```

```python
import re
pro['text'] = pro['text'].apply(lambda x:re.sub(r"[@#$%^""&*’(_.)-+=!?/|\\',;:]", '',x) if x==x else x)
pro['text'] = pro['text'].apply(lambda x:re.sub(r"[“’”]", '',x))
pro['text'] = pro['text'].apply(lambda x:re.sub(" \d+", "", x))
```

```python
pro['text']
```

```python
import nltk
import swifter
from rake_nltk import Rake
nltk.download('stopwords')
nltk.download('punkt')
r = Rake()

def rakeprocessing(text):
    r.extract_keywords_from_text(text)
    return r.get_ranked_phrases_with_scores()
```

```python
pro['Rake'] = pro.text.apply(lambda x:rakeprocessing(x))
```

```python
pro['Rake']
```

```python
rke = pro.explode('Rake')
print(rke)
rke=rke.dropna(subset=['Rake'])
rke['Rake_text'] = rke.Rake.apply(lambda x:x[1])
rke['Rake_Score'] = rke.Rake.apply(lambda x:str(int(x[0])))
```

```python
rke = rke[rke['Rake_Score'].isin(['1','2','3','4'])]
```

```python
rke = rke.drop(['Rake'],axis=1)
```

```python
rke
```

```python
rke['image_url'] = ['https://www.instagram.com/p/{}/media/?size=l'.format(x) for x in rke['shortcode']]
```

```python
from transformers import pipeline
classifier = pipeline('sentiment-analysis')
def sentimentprocessing(text):
    try:
        return classifier(text)
    except Exception as e:
        print(e)
        return text

import swifter
sen['Sentiment_Analysis'] = sen.text.swifter.apply(lambda x:sentimentprocessing(x))
sen['Sentiment_Score'] = sen['Sentiment_Analysis'].apply(lambda x:x[0]['score'])
sen['Sentiment'] = sen['Sentiment_Analysis'].apply(lambda x:x[0]['label'])
```

```python
sen['image_url'] = ['https://www.instagram.com/p/{}/media/?size=l'.format(x) for x in sen['shortcode']]
```
